```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:/Users/User/OIBSIP/RetailSalesEDA/data/raw/Warehouse_and_Retail_Sales.csv')

# Fill missing values in 'SUPPLIER' with 'Unknown'
data['SUPPLIER'].fillna('Unknown', inplace=True)

# Drop the row with missing value in 'ITEM TYPE'
data.dropna(subset=['ITEM TYPE'], inplace=True)

# Impute missing values in 'RETAIL SALES' with the median
median_retail_sales = data['RETAIL SALES'].median()
data['RETAIL SALES'].fillna(median_retail_sales, inplace=True)

# Calculate descriptive statistics for numerical columns
descriptive_stats = data[['RETAIL SALES', 'RETAIL TRANSFERS', 'WAREHOUSE SALES']].describe()

# Calculate mode for numerical columns
mode_retail_sales = data['RETAIL SALES'].mode()[0]
mode_retail_transfers = data['RETAIL TRANSFERS'].mode()[0]
mode_warehouse_sales = data['WAREHOUSE SALES'].mode()[0]

# Add mode to the descriptive statistics
descriptive_stats.loc['mode'] = [mode_retail_sales, mode_retail_transfers, mode_warehouse_sales]

# Display the descriptive statistics
print("Descriptive Statistics:")
print(descriptive_stats)

# Aggregate sales data by year and month
monthly_sales = data.groupby(['YEAR', 'MONTH'])[['RETAIL SALES', 'WAREHOUSE SALES']].sum().reset_index()

# Create a date column for better visualization
monthly_sales['DATE'] = pd.to_datetime(monthly_sales[['YEAR', 'MONTH']].assign(DAY=1))

# Plot retail sales and warehouse sales over time
plt.figure(figsize=(14, 7))
plt.plot(monthly_sales['DATE'], monthly_sales['RETAIL SALES'], label='Retail Sales')
plt.plot(monthly_sales['DATE'], monthly_sales['WAREHOUSE SALES'], label='Warehouse Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Trends Over Time')
plt.legend()
plt.show()

# Group data by item type and calculate total sales
item_type_sales = data.groupby('ITEM TYPE')['RETAIL SALES'].sum().reset_index()

# Plot total sales by item type
plt.figure(figsize=(10, 6))
plt.bar(item_type_sales['ITEM TYPE'], item_type_sales['RETAIL SALES'], color='skyblue')
plt.xlabel('Item Type')
plt.ylabel('Total Retail Sales')
plt.title('Total Retail Sales by Item Type')
plt.xticks(rotation=45)
plt.show()

# Group data by supplier and calculate total sales
supplier_sales = data.groupby('SUPPLIER')['RETAIL SALES'].sum().reset_index()

# Get top 10 suppliers by retail sales
top_suppliers = supplier_sales.nlargest(10, 'RETAIL SALES')

# Plot top 10 suppliers
plt.figure(figsize=(12, 6))
plt.bar(top_suppliers['SUPPLIER'], top_suppliers['RETAIL SALES'], color='green')
plt.xlabel('Supplier')
plt.ylabel('Total Retail Sales')
plt.title('Top 10 Suppliers by Retail Sales')
plt.xticks(rotation=45)
plt.show()

# Calculate correlation matrix
corr_matrix = data[['RETAIL SALES', 'RETAIL TRANSFERS', 'WAREHOUSE SALES']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
