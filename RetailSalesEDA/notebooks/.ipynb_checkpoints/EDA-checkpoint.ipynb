```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:/Users/User/RetailSalesEDA/data/raw/menu.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Display the data types
print("\nData Types:\n", data.info())

# List of nutritional columns
nutritional_columns = [
    'Calories', 'Calories from Fat', 'Total Fat', 'Total Fat (% Daily Value)',
    'Saturated Fat', 'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol',
    'Cholesterol (% Daily Value)', 'Sodium', 'Sodium (% Daily Value)',
    'Carbohydrates', 'Carbohydrates (% Daily Value)', 'Dietary Fiber',
    'Dietary Fiber (% Daily Value)', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)',
    'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'
]

# Descriptive statistics for nutritional columns
descriptive_stats = data[nutritional_columns].describe()
print("\nDescriptive Statistics:\n", descriptive_stats)

# Calculate category-wise nutritional averages
category_means = data.groupby('Category')[nutritional_columns].mean().reset_index()

# Plot the average nutritional content for each category
plt.figure(figsize=(20, 15))
for i, column in enumerate(nutritional_columns, 1):
    plt.subplot(6, 4, i)
    sns.barplot(data=category_means, x='Category', y=column)
    plt.xticks(rotation=90)
    plt.title(f'Average {column} by Category')
plt.tight_layout()
plt.show()

# Plot distribution of Calories, Total Fat, Sugars, and Protein by Category
plt.figure(figsize=(20, 10))
for i, column in enumerate(['Calories', 'Total Fat', 'Sugars', 'Protein'], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=data, x='Category', y=column)
    plt.xticks(rotation=90)
    plt.title(f'Distribution of {column} by Category')
plt.tight_layout()
plt.show()
```
