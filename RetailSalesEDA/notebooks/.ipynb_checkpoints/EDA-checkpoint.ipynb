```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:/Users/User/OIBSIP/Retails-sales-EDA/data/raw/menu.csv'
data = pd.read_csv(file_path)


print(data.head())

# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Display the data types
print("\nData Types:\n", data.info())

# List of nutritional columns
nutritional_columns = [
    'Calories', 'Calories from Fat', 'Total Fat', 'Total Fat (% Daily Value)',
    'Saturated Fat', 'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol',
    'Cholesterol (% Daily Value)', 'Sodium', 'Sodium (% Daily Value)',
    'Carbohydrates', 'Carbohydrates (% Daily Value)', 'Dietary Fiber',
    'Dietary Fiber (% Daily Value)', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)',
    'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'
]

# Descriptive statistics for nutritional columns
descriptive_stats = data[nutritional_columns].describe()
print("\nDescriptive Statistics:\n", descriptive_stats)

# Calculate category-wise nutritional averages
category_means = data.groupby('Category')[nutritional_columns].mean().reset_index()

# Plot the average nutritional content for each category
fig, axes = plt.subplots(len(nutritional_columns), 1, figsize=(20, 150))
for i, column in enumerate(nutritional_columns):
    sns.barplot(data=category_means, x='Category', y=column, ax=axes[i])
    axes[i].set_title(f'Average {column} by Category')
    axes[i].set_xticks(range(len(category_means['Category'])))
    axes[i].set_xticklabels(category_means['Category'], rotation=90)
plt.subplots_adjust(hspace=1.0)
plt.show()

# Plot distribution of Calories, Total Fat, Sugars, and Protein by Category
fig, axes = plt.subplots(4, 1, figsize=(20, 40))
for i, column in enumerate(['Calories', 'Total Fat', 'Sugars', 'Protein']):
    sns.boxplot(data=data, x='Category', y=column, ax=axes[i])
    axes[i].set_title(f'Distribution of {column} by Category')
    axes[i].set_xticks(range(len(data['Category'].unique())))
    axes[i].set_xticklabels(data['Category'].unique(), rotation=90)
plt.subplots_adjust(hspace=0.5)
plt.show()

# Correlation matrix
corr_matrix = data[nutritional_columns].corr()
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Nutritional Values")
plt.show()

# Recommendations based on EDA
recommendations = """
1. **Focus on Calorie Control**:
   - **High-Calorie Items**: Consider offering lower-calorie alternatives for high-calorie items, particularly in the categories with higher averages (e.g., Desserts, Beef & Pork).
   - **Portion Control**: Reduce portion sizes or offer smaller portion options to help manage calorie intake.

2. **Fat and Saturated Fat Reduction**:
   - **Healthy Fats**: Replace saturated fats with healthier fats (e.g., from avocados, nuts, or olive oil).
   - **Menu Redesign**: Revise recipes to lower the total and saturated fat content, especially in high-fat categories like Beef & Pork and Chicken & Fish.

3. **Sodium Reduction**:
   - **Lower Sodium Alternatives**: Offer menu items with reduced sodium content, particularly in categories with high sodium averages (e.g., Beef & Pork).
   - **Ingredient Substitution**: Use herbs, spices, and other flavor enhancers instead of salt.

4. **Increase Dietary Fiber**:
   - **Fiber-Rich Ingredients**: Incorporate more fiber-rich ingredients such as whole grains, fruits, vegetables, and legumes in the menu items.
   - **Promote Fiber Benefits**: Educate customers on the benefits of dietary fiber and highlight fiber-rich options on the menu.

5. **Balanced Nutritional Profile**:
   - **Nutritional Balance**: Ensure a balanced distribution of macronutrients (carbohydrates, proteins, and fats) across all menu categories.
   - **Customizable Meals**: Provide options for customers to customize their meals to meet their nutritional needs and preferences.

6. **Transparency and Education**:
   - **Nutritional Information**: Make nutritional information easily accessible to customers to help them make informed choices.
   - **Healthy Eating Campaigns**: Launch campaigns to promote healthy eating habits and inform customers about the nutritional content of menu items.

7. **Innovation and Variety**:
   - **New Recipes**: Innovate and introduce new recipes that cater to health-conscious customers while maintaining flavor and satisfaction.
   - **Seasonal Menus**: Offer seasonal menus that include fresh, seasonal ingredients, which can help in providing variety and nutrition.

8. **Customer Feedback**:
   - **Surveys and Feedback**: Regularly collect feedback from customers regarding their preferences and perceptions about the nutritional quality of the menu items.
   - **Adaptation**: Use the feedback to continuously adapt and improve the menu offerings.

By implementing these recommendations, the restaurant can cater to a health-conscious customer base, promote better eating habits, and potentially increase customer satisfaction and loyalty.
"""

print("Recommendations:\n", recommendations)
```
